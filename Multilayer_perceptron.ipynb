```python
import tensorflow as tf
import numpy as np

X = np.array([
    [2, 8],  # Fail
    [4, 6],  # Fail
    [6, 7],  # Pass
    [8, 5],  # Pass
    [10, 4], # Pass
    [1, 9]   # Fail
], dtype=np.float32)

y = np.array([0, 0, 1, 1, 1, 0], dtype=np.float32)

```

```python
# Step 2: Create a TensorFlow model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='sigmoid', input_shape=(2,)),  # First hidden layer with 8 neurons
    tf.keras.layers.Dense(4, activation='sigmoid'),                   # Second hidden layer with 4 neurons
    tf.keras.layers.Dense(1, activation='sigmoid')                # Output layer with 1 neuron (binary classification)
])

# Step 3: Compile the modely
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```python
model.summary()
```

```python
# Step 4: Train the model
model.fit(X, y, epochs=100, verbose=1)

```

```python
# Step 5: Test the model
test_data = np.array([
    [3, 7],  # Likely Fail
    [7, 6],  # Likely Pass
    [9, 4]   # Likely Pass
], dtype=np.float32)
```

```python

predictions = model.predict(test_data)

# Output results
for i, test in enumerate(test_data):
    print(f"Input: {test}, Prediction: {'Pass' if predictions[i] > 0.5 else 'Fail'}")
```
